```python
import numpy as np

# Data yang diberikan
x = np.array([1.1, 1.3, 1.5, 2, 2.2, 2.9, 3, 3.2, 3.2, 3.7])
y = np.array([39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189])

# Hitung sigma () untuk x, y, xy, dan x^2
sum_x = np.sum(x)
sum_y = np.sum(y)
xy = x * y  # Perkalian elemen-wise
sum_xy = np.sum(xy)
x_squared = x**2  # Kuadrat elemen-wise
sum_x_squared = np.sum(x_squared)

# Output
print("Langkah 1: x =", sum_x)
print("Langkah 2: y =", sum_y)
print("Langkah 3: xy =", xy)
print("Langkah 4: xy =", sum_xy)
print("Langkah 5: x^2 =", x_squared)
print("Langkah 6: x^2 =", sum_x_squared)

# Hitung m dan b
n = len(x)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
b = (sum_y - m * sum_x) / n

# Output
print("\nPerhitungan m dan b:")
print("Langkah 7: n =", n)
print(f"Langkah 8: m = ({n} * {sum_xy} - {sum_x} * {sum_y}) / ({n} * {sum_x_squared} - {sum_x}^2)")
print("Langkah 9: m =", m)
print("Langkah 10: b = (y - mx) / n")
print(f"Langkah 11: b = ({sum_y} - {m} * {sum_x}) / {n}")
print("Langkah 12: b =", b)

# Hitung prediksi gaji untuk lama kerja 4 tahun
lama_kerja_baru = 4
gaji_prediksi = m * lama_kerja_baru + b

# Output
print("\nPrediksi gaji untuk lama kerja 4 tahun:")
print(f"Langkah 13: Prediksi gaji = m * {lama_kerja_baru} + b")
print("Langkah 14: Prediksi gaji =", gaji_prediksi)
```
